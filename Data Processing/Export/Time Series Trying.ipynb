```python
data = []
for i in range(0, len(years)):
    internal = []
    index = 0
    df = pd.read_csv(f"../../Datasets/Export/Export-Product-Wise/{years[i]}.csv")

    for j in range(0, len(df) - 1):
        if df["Product Name"].str.contains("FLORICULTURE")[j] == 1:
            index = j
            break

    internal.append(years[i])
    internal.append(df.iloc[index]["Qty(MT)"])
    internal.append(df.iloc[index]["Rs(Crore)"])
    internal.append(df.iloc[index]["% Share(Qty)"])
    internal.append(df.iloc[index]["% Share(Rs)"])
    data.append(internal)

df1 = pd.DataFrame(data, columns=["Year", "Qty(MT)", "Rs(Crore)", "% Share(Qty)", "% Share(Rs)"])
df1.loc[len(df1.index)] = ["Total", df1["Qty(MT)"].sum(), df1["Rs(Crore)"].sum(), 0, 0]
df1.to_csv("../../Datasets/Export/Product-Wise-Time-Series/Floriculture.csv", index=None)
```

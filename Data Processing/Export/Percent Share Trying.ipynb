```python
totalQty = float((df.iloc[-1]["Qty(MT)"]).replace(",", ""))
totalRs = float((df.iloc[-1]["Rs(Crore)"]).replace(",", ""))
totalQty, totalRs
```

```python
type(df.iloc[-1]["Qty(MT)"])
```

```python
df["Qty(MT)"] = df["Qty(MT)"].apply(lambda x: x.replace(",", "")).astype(float)
df["Rs(Crore)"] = df["Rs(Crore)"].apply(lambda x: x.replace(",", "")).astype(float)
```

```python
df["% Share(Qty)"] = ((df["Qty(MT)"] / totalQty) * 100).round(3)
df["% Share(Rs)"] = ((df["Rs(Crore)"] / totalRs) * 100).round(3)
```

```python
df
```

```python
df.to_csv("../../Datasets/Export/Export-Product-Wise/2021-22.csv")
```
